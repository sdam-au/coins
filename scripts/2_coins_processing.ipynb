```python
def clean_date(datetime_string):
    date = None
    if len(datetime_string) >=10:
        date = datetime_string.partition("-")[0]
    else:
        date = datetime_string
    if date[:2] in ["19", "20"]:
        date = "00" + date[2:]
    try:
        date = int(date)
    except:
        date = None
    return date
coins["not_before"] = coins["ontology#hasStartDate"].apply(clean_date)
coins["not_after"] = coins["ontology#hasEndDate"].apply(clean_date)
```
